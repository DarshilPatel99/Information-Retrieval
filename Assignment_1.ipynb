### Import necessary libraries

*   `nltk` is used to perform NLP operations
*   `BeautifulSoup` is used to read html or xml files

```
import os
import re
import nltk as nk
from bs4 import BeautifulSoup as bs

#installing useful data of nltk for preprocessing
nk.download('stopwords')
nk.download('wordnet')
nk.download('punkt')
nk.download('averaged_perceptron_tagger')
```

###Dataset

*   get the dataset(corpus) in xml format which contains text in `<TEXT>` tag.  

```
#path of data files
data_dir = '/content/drive/MyDrive/IR LAB/Assignment 1/data/'
data_files = os.listdir(data_dir)
print(data_files)
```

###Read and combine `<TEXT>` tag's data 

*   `beautifulSoupObject.find()` method will extract text from `<TEXT>` tag.

```
text_data = []

for i in data_files:
  with open(os.path.join(data_dir,i),'r') as f:
    content = f.readlines()
    content = "".join(content)
    bs_content = bs(content, "lxml")
    temp_text_data = bs_content.find('text').text
    text_data.append(temp_text_data)

my_text = ' '.join(text_data)
print(my_text)
```

###Remove punctuation

*   `re.sub()` function is used to remove punctuation marks from text.

```
my_text2 = re.sub(r'[^\w\s]', '', my_text)
print(my_text2)
```

###Remove numerical values

*   `re.sub()` function is used to remove numerical values from text.

```
#removes numerical value(word) from text
my_text3 = re.sub(r'\b\d+\b', '', my_text2)

#removes all degites from the text
#my_text3 = re.sub(r'\d+', '', my_text2)

print(my_text3)
```

###Tokenization

*   There are two methods for Tokenization
    1.   Using `stringObject.split()`
    2.   Using `nltk.word_tokenize()`

*   If text does not contain any punctuation marks, then both method works same otherwise 1st method separate all tokens by space and 2nd method separate all tokens by any word separator and counts punctuation marks as separate token.

```
#similarity between both method......

#tokenization using string's function
tokens_split = my_text3.split()
print(len(tokens_split))
print(tokens_split)

#tokenization using nltk
tokens_nltk = nk.word_tokenize(my_text3)
print(len(tokens_nltk))
print(tokens_nltk)

#check both token lists are same or not
if tokens_nltk == tokens_split:
  print(True)
else:
  print(False)
```

```
#difference between both method......

#tokenization using string's function
tokens_split1 = my_text.split()
print(len(tokens_split1))
print(tokens_split1)

#tokenization using nltk
tokens_nltk1 = nk.word_tokenize(my_text)
print(len(tokens_nltk1))
print(tokens_nltk1)

#check both token lists are same or not
if tokens_nltk1 == tokens_split1:
  print("True")
else:
  print("False")
```

###Converting text to lowercase

*   `str.lower()` function is used to convert all tokens to lowercase.



```
#Convert tokens to lowercase----------------------

tokens_lower = list(map(str.lower,tokens_nltk))
print(len(tokens_lower))
print(tokens_lower)
```

###Remove stop-words

*   `nltk` has inbuilt corpus of stopwords so remove those words from token list.


```
#Remove stop-words-------------------------------

stopwords = set(nk.corpus.stopwords.words('english'))
print(len(stopwords))
print(stopwords)

tokens_non_stopwords = [token for token in tokens_lower if token not in stopwords]
print(len(tokens_non_stopwords))
print(tokens_non_stopwords)
```

###Stemming

*   `PorterStrmmer` is used to perform Stemming operation.
*   In stemming process it removes suffix from the word and try to convert it into root word. But many times it gives incorrect root word because of the rules of grammer in language.    

```
#Stemming----------------------------------------

stemmer = nk.stem.PorterStemmer()
stemmed_tokens = list(map(stemmer.stem,tokens_non_stopwords))
print(len(stemmed_tokens))
print(stemmed_tokens)
```

###Lemmatization

*   `nltk.stem.WordNetLemmatizer()` is used to perform Lemmatization.
*   Lemmatization is also used to find the the root word but it uses vocabulary and morphological analysis of words instead of removing suffix. so, it gives accurate result.

```
#Lemmatization----------------------------------------
lemmatizer = nk.stem.WordNetLemmatizer()
lemmatized_tokens = list(map(lemmatizer.lemmatize,tokens_non_stopwords))
print(len(lemmatized_tokens))
print(lemmatized_tokens)

#part of speech tag(POS tag)----------------------------------
#token_pos_tag = dict(nk.pos_tag(token_non_stopwords))
#print(token_pos_tag)
```

###Difference between Stemming and Lemmatization

*   Stemming looks like hard codded and removes exctra suffix letters from word to get root word while Lemmatization take care of grammer and morphological structure of word to get root word.

```
#difference between Stemming and Lemmatization
print([(word,stemmed_word,lemmatized_word) for word,stemmed_word,lemmatized_word in zip(tokens_non_stopwords,stemmed_tokens,lemmatized_tokens)])
```
